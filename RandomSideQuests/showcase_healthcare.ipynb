```python
import pandas as pd
import numpy as np

health_df = pd.DataFrame({
    "patient_id": [101,102,103,104,105,101,102,103,104,105,106,107,108,109],
    "patient_name": ["Alice","Bob","Charlie","Diana","Eve","Alice","Bob","Charlie","Diana","Eve","Frank","Grace","Hank","Ivy"],
    "age": [34, 45, 29, 50, 38, 34, 45, 29, 50, 38, 60, 42, 31, 27],
    "gender": ["F","M","M","F","F","F","M","M","F","F","M","F","M","F"],
    "visit_date": pd.to_datetime([
        "2024-03-01","2024-03-02","2024-03-02","2024-03-03","2024-03-04",
        "2024-03-10","2024-03-11","2024-03-11","2024-03-12","2024-03-12",
        "2024-03-15","2024-03-15","2024-03-16","2024-03-16"
    ]),
    "systolic_bp": [120, 140, 130, 150, 135, 125, 138, 128, 148, 132, 160, 145, 118, np.nan],
    "diastolic_bp": [80, 90, 85, 95, 88, 82, 87, 83, 92, 86, 100, 94, 77, 80],
    "heart_rate": [72, 88, 76, 95, 80, 74, 85, 78, 90, 82, 98, 89, 70, 75],
    "cholesterol_mg_dl": [190, 220, 180, 240, 210, 185, 215, 175, 235, 205, 250, 225, 170, 180],
    "diagnosis": ["Healthy","Hypertension","Healthy","Hypertension","Prehypertension",
                  "Healthy","Hypertension","Healthy","Hypertension","Prehypertension",
                  "Hypertension","Prehypertension","Healthy","Healthy"]
})

# Introduce some duplicates for realism
health_df = pd.concat([health_df, health_df.iloc[[1,4]]], ignore_index=True)
health_df
```

```python
'''
Task 1 - Clean data
'''

df_cleaned = (
    health_df
    .drop_duplicates()
    .assign(systolic_bp = lambda x: x.systolic_bp.fillna(x.systolic_bp.median()))
)
df_cleaned
```

```python
'''
Task 2 - Compute new metrics
'''
def calculate_bp_category(systolic_bp_value):
    if systolic_bp_value < 120:
        return "Normal"
    elif systolic_bp_value < 140:
        return "Prehypertension"
    else:
        return "Hypertension"

df_metrics = (
    df_cleaned
    .assign(mean_arterial_pressure = (2*df_cleaned.diastolic_bp + df_cleaned.systolic_bp)/3)
    .assign(bp_category = lambda x: x.systolic_bp.apply(lambda y: calculate_bp_category(y)))
)
df_metrics
```

```python
'''
Task 3 - Latest Visit per Patient
'''
df_patient_latest_visit = (
    df_metrics
    .loc[df_metrics.groupby('patient_id')['visit_date'].idxmax()][['patient_id','visit_date','diagnosis','mean_arterial_pressure']]
)
df_patient_latest_visit
```

```python
'''
Task 4 - Gender-based analysis
'''
df_gender_analysis = (
    df_cleaned
    .groupby('gender')
    .agg(
        avg_systolic_bp = ('systolic_bp','mean'),
        avg_diastolic_bp = ('diastolic_bp','mean'),
        avg_heart_rate = ('heart_rate','mean'),
        number_of_visits = ('patient_id','count')
    )
)
df_gender_analysis
```

```python
'''
Task 5 - High risk patients
'''
df_high_risk_patients = (
    df_cleaned
    .loc[(df_cleaned.systolic_bp >=140) & (df_cleaned.cholesterol_mg_dl >=230)]
    .sort_values('systolic_bp',ascending=False)
)
df_high_risk_patients
```

```python
'''
Task 6 - Time based analysis
'''
df_time_analysis = (
    df_cleaned
    .groupby(pd.Grouper(key='visit_date', freq='W'))['heart_rate']
    .mean()
    .to_frame()
)
df_time_analysis
```
